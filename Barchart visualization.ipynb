```python
countries = pd.read_csv("/content/countries.csv")
df = countries.copy()
```

```python
df.head(3)
```

```python
#comparing the population growth between Us and china
```

```python
us = df[df.country=="United States"]
us.value_counts().sum()
```

```python
china = df[df.country=="China"]
china.value_counts().sum()
```

```python
#visualization
plt.plot(us.year,us.population/10**6)
plt.plot(china.year,china.population/10**6)
plt.legend(["US","China"])
plt.xlabel("year")
plt.ylabel("population")
plt.show()
```

```python
us.population
```

```python
#dividing the all year us popualtion with 1608 year
round(us.population/us.population.iloc[0]*100,2)
```

```python
#dividing the all year population of china with 1608 year
round(china.population/china.population.iloc[0]*100,2)
```

```python
plt.plot(us.year,round(us.population/10**6,2))
plt.plot(china.year,round(china.population/10**6,2))
plt.legend(["US","China"])
plt.xlabel("year")
plt.ylabel("population")
plt.show()
```

```python
#comparing both the us and china population of 1608
plt.plot(us.year,us.population/us.population.iloc[0]*100)
plt.plot(china.year,china.population/china.population.iloc[0]*100)
plt.legend(["US","China"])
plt.xlabel("year")
plt.ylabel("population growth(first year==100)")
plt.show()
```

```python
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
blood_sugar_men = [77, 80, 82, 85, 88, 90, 93, 113, 115, 129, 135, 149, 150]
```

```python
plt.xlabel("Sugar range")
plt.ylabel("Total no of patients")
plt.title("Blood Sugar Analysis")
plt.hist([blood_sugar_women,blood_sugar_men ],bins=[80,100,125,150],rwidth=0.90,color=["purple","pink"],label=["women","men"]) #rwidth is used so that it doesnt attach with other graph
plt.legend()                                                                                                                  #orientation makes your histogram in horizontal fashion                              
plt.show()
```
