```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
#Axes labels, Legend, Grid
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.plot(days,max_t,label="max")
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label = "avg")
plt.title("Weather")
#plt.legend(["max temp","min temp","avg temp"])
plt.legend(loc="best",shadow=True,fontsize="large") #upper right #upper left #best find the blank area and insert it into best positio
plt.grid()
plt.show()
```

Simple bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

Vertical Bar chart

```python
xpos = np.arange(len(company))
plt.xticks(xpos,company) #connect element with coresponding   
plt.title("Us Tech stocks")
plt.bar(xpos-0.2,revenue,width=0.4,) #barchart for revenue
plt.bar(xpos+0.2,profit,width=0.4,) #barchart for profit
plt.legend(["revenue","profit"])
plt.show()
```

Horizontal Bar chart

```python
#using arange so that we can we can edit barchart multiple way
xpos = np.arange(len(company))
plt.yticks(xpos,company) #it compare element horizontal
#represent title in barchart 
plt.title("Us Tech stocks")
plt.barh(xpos-0.2,revenue,height=0.4,label="revenue") #use height avoid overlapping while                                                 
plt.barh(xpos+0.2,profit,height=0.4,label="profit") #plotting the bar graph horizontally
plt.legend()
plt.show()
```
